# <p style="color:Blue;"><center>Enfermedades respiratorias</center></p>
![imagen](https://www.yoinfluyo.com/images/stories/hoy/nov20/301120/influenza_salud.jpg)

<h2><center>Introduccion</center></h2>
El adulto promedio respira de 15 a 20 veces por minuto, más de 20,000 respiraciones por día. Los pulmones son diferentes a la mayoría de los órganos de su cuerpo, ya que están conectados directamente al ambiente exterior. Cualquier cosa que inhala puede afectar a los pulmones.
Su cuerpo tiene un sistema de defensa natural diseñado para proteger a los pulmones. También puede reducir su riesgo de desarrollar enfermedades pulmonares si no fuma, si evita la exposición al humo de segunda mano, si evita la exposición a la contaminación del aire, los productos químicos y otros irritantes, y si se realiza chequeos regulares de salud (entre ellos, las vacunas).

## ¿Qué son las enfermedades respiratorias?

Es un tipo de enfermedad que afecta los pulmones y otras partes del aparato respiratorio. Las enfermedades respiratorias se producen por infecciones, consumo de tabaco o inhalación de humo de tabaco en el ambiente, y exposición al radón, amianto u otras formas de contaminación del aire. Las enfermedades respiratorias incluyen el asma, la enfermedad pulmonar obstructiva crónica (EPOC), la fibrosis pulmonar, la neumonía y el cáncer de pulmón. También se llama enfermedad pulmonar y trastorno pulmonar.

<h2>Algunas de las principales enfermedades respiratorias:</h2>

<h4>Asma</h4>

El asma dificulta la respiración y afecta a millones de estadounidenses de todas las edades. No existe una cura para el asma, pero puede controlarse para que pueda vivir una vida sana y activa. [Obtenga más información sobre el asma.](https://www.lung.org/espanol/salud-pulmonar-y-enfermedades/asma)

<h4>Enfermedad pulmonar obstructiva crónica (EPOC)</h4>

La enfermedad pulmonar obstructiva crónica (EPOC), también conocida como enfisema y bronquitis crónica, es la cuarta causa principal de muerte en los Estados Unidos. La buena noticia es que la EPOC es a menudo prevenible y tratable.[Obtenga más información sobre la EPOC.](https://www.lung.org/espanol/salud-pulmonar-y-enfermedades/epoc)

<h4>Influenza (gripe)</h4>

La influenza (gripe) es una infección viral altamente contagiosa y es una de las enfermedades más graves de la temporada de invierno. La influenza se transmite fácilmente de persona a persona, generalmente cuando una persona infectada tose o estornuda. La mejor forma de protegerse contra la gripe es vacunarse todos los años.[Obtenga más información sobre la gripe.](https://www.lung.org/espanol/salud-pulmonar-y-enfermedades/influenza)

<h4>Cáncer de pulmón</h4>

Mueren más personas de cáncer de pulmón que de cualquier otro tipo de cáncer. Aunque el cáncer de pulmón es común, muchas personas no saben mucho acerca de él o tienen miedo de informarse. Entender el cáncer de pulmón es un paso importante en la prevención de las muertes por cáncer de pulmón.[Obtenga más información sobre el cáncer de pulmón.](https://www.lung.org/espanol/salud-pulmonar-y-enfermedades/cancer-de-pulmon)

<h4>Neumonía</h4>

La neumonía es una infección en uno de sus pulmones, o en ambos, que causa inflamación y acumulación de líquido. Cualquier persona puede contraer neumonía, pero algunas personas tienen un riesgo más alto que las demás.[Obtenga más información sobre la neumonía.](https://www.lung.org/espanol/salud-pulmonar-y-enfermedades/neumona)

<h4>Fibrosis pulmonar</h4>

La fibrosis pulmonar (FP) es una enfermedad en la que hay cicatrices en los pulmones que dificultan la respiración. Hay muchos tipos diferentes de fibrosis pulmonar. El más común es la fibrosis pulmonar idiopática o FPI, que no tiene causa conocida. [Obtenga más información sobre la fibrosis pulmonar.](https://www.lung.org/espanol/salud-pulmonar-y-enfermedades/fibrosis-pulmonar)

<h2>Síntomas de las enfermedades respiratorias</h2>

Los síntomas varían según la persona y según la patología. No obstante, los síntomas más frecuentes son:

- Tos
- Silbidos y presión en el pecho
- Secreciones (flemas)
- Fatiga y dificultad para respirar
- Rinitis
- Estornudos
- Picor de nariz

<h2>¿Cuáles son las causas de las enfermedades respiratorias?</h2>
Puede haber combinación de factores que causen las enfermedades respiratorias. Normalmente responden a determinadas causas:

- Virus
- Fumar cigarrillos
- contaminación del aire, polvo y gases tóxicos

Algunos factores de riesgo son:

- El humo del tabaco
- Baja resistencia del sistema inmunológico
- Exposición a agentes irritantes en el lugar de trabajo
- Reflujo gástrico

<h2>¿Se pueden prevenir?</h2>

Aunque actualmente las enfermedades respiratorias no se pueden prevenir totalmente, sí se pueden evitar algunos de los problemas asociados. Así, se recomienda proteger los pulmones lo máximo posible siguiendo este tipo de consejos:

- Evitar el humo del tabaco: el humo aumenta el riesgo de padecer algún tipo de enfermedad respiratoria.
- Vacunarse: muchas enfermedades respiratorias se manifiestan a raíz de un virus, existen vacunas para evitar problemas mayores.
- Lavarse las manos: la higiene en las manos ayudará a los pacientes a mantener la higiene y así evitar el riesgo de contraer una infección viral.
- Usar mascarilla quirúrgica: la utilización de este aparato puede evitar complicaciones en los trabajos en los que el paciente se expone al polvo, vapores o gases tóxicos.
