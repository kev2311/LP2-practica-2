```python
# Importando las librer√≠as
import json
import pandas as pd
import numpy as np
import seaborn as sns
from urllib.request import urlopen
from bs4 import BeautifulSoup as soup
import requests
```

```python
# Mercado libre - celulares
# Accediendo al website
url="https://listado.mercadolibre.com.pe/computacion/celulares"
uclient=urlopen(url)
pagehtml=uclient.read()
uclient.close()
pagesoup=soup(pagehtml, "html.parser")
c=pagesoup.findAll("li",{"class":"ui-search-layout__item"}) #para la busqueda de cada articulo(celulares)
c[1].div.a["title"]
```

```python
#Insertamos las listas vacias de lo que queremos extraer
titulos=[]
precio_actual=[]
descuento=[]
precio_anterior=[]

#find busca un elemento
#findAll crea una lista

for i in range(0,4): #Queremos  
    url="https://listado.mercadolibre.com.pe/computacion/celulares/"+"_Desde_"+str(i*50+1) 
    uclient=urlopen(url)
    pagehtml=uclient.read()  #leer el url
    uclient.close()
    pagesoup=soup(pagehtml, "html.parser")
    c=pagesoup.findAll("li",{"class":"ui-search-layout__item"})  #para scrapear cada producto

    for i in range(0,len(c)): #de lo scrapeado
        
        titulo=c[i].div.a["title"]
        #mas especifico         
        titulo=c[i].find("div",{"ui-search-item__group ui-search-item__group--title"}).a["title"] #[0] ya que guarda todo en una lista 
                                                                                                  #de un elemento y agarramos ese element                
        titulos.append(titulo)
        #titulos.append(titulo)
        if len(c[i].findAll("span",{"price-tag-fraction"})) == 2:
           
            precioactual=c[i].findAll("span",{"price-tag-fraction"})[1].text
            precioanterior=c[i].findAll("span",{"price-tag-fraction"})[0].text
            descuento1=c[i].find("span",{"class":"ui-search-price__discount"}).text
            precio_actual.append(precioactual)
            precio_anterior.append(precioanterior)
            descuento.append(descuento1)

        elif len(c[i].findAll("span",{"price-tag-fraction"})) == 1:
            descuento1=0
            descuento.append(descuento1)
            precioactual=c[i].findAll("span",{"price-tag-fraction"})[0].text
            precioanterior=c[i].findAll("span",{"price-tag-fraction"})[0].text
            precio_actual.append(precioactual)
            precio_anterior.append(precioanterior)


        
            
for i in range(0,len(precio_actual)):
    precio_actual[i]=float(precio_actual[i].replace(".","")) #3599.0
    #precio_actual[i]=float(precio_actual[i])  #3.599
    precio_anterior[i]=float(precio_anterior[i].replace(".",""))
```

```python
# Visualizando la data con formato Json
data ={"Nombre":titulos,"Precio actual":precio_actual,"Precio anterior":precio_anterior
                 ,"Descuento":descuento}
#Nos devuelve el String con el JSON
data_string = json.dumps(data)
print ('JSON:', data_string)
```

```python
#Creando Data Frame
df=pd.DataFrame({"Nombre":titulos,"Precio actual":precio_actual,"Precio anterior":precio_anterior
                 ,"Descuento":descuento},index=list(range(1,len(titulos)+1))) #5titulos

#Pasando lo datos a un csv
df.to_csv("celulares.csv")
```
